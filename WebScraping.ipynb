### Retrieving the content of the articles

```python
#lets create a function that allows us to enter the keyword we want to search 
def get_titles(keywords):
    all_news = []
    # Create a GoogleNews object with French language and country set to France
    gn = GoogleNews(lang='fr', country='FR')
     # Loop through each keyword in the list
    for keyword in keywords:
        search = gn.search(keyword)
        articles = search['entries']
         # Loop through each article in the list
        for article in articles:
             # Create a dictionary to store article data
            article_data = {
                'title': article.title,
                'link': article.link,
                'published': article.published,
                'Category': keyword
            }
            # Append the article data to the list of all news
            all_news.append(article_data)
            
    return all_news



# Get news data for all keywords
data = get_titles(keywords)
print(data)

```

### DataFrame of the collected data

```python

df = pd.DataFrame(data)
len(df)
```
